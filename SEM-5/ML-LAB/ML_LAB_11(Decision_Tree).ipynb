```python
# Import necessary libraries
from sklearn import tree
from sklearn.preprocessing import OneHotEncoder
from tabulate import tabulate
import matplotlib.pyplot as plt

# Sample data
data = [
    ['Sunny', 'Hot', 'High', 'Weak', 'No'],
    ['Sunny', 'Hot', 'High', 'Strong', 'No'],
    ['Overcast', 'Hot', 'High', 'Weak', 'Yes'],
    ['Rainy', 'Mild', 'High', 'Weak', 'Yes'],
    ['Rainy', 'Cool', 'Normal', 'Weak', 'Yes'],
    ['Rainy', 'Cool', 'Normal', 'Strong', 'No'],
    ['Overcast', 'Cool', 'Normal', 'Strong', 'Yes'],
    ['Sunny', 'Mild', 'High', 'Weak', 'No'],
    ['Sunny', 'Cool', 'Normal', 'Weak', 'Yes'],
    ['Rainy', 'Mild', 'Normal', 'Weak', 'Yes'],
    ['Sunny', 'Mild', 'Normal', 'Strong', 'Yes'],
    ['Overcast', 'Mild', 'High', 'Strong', 'Yes'],
    ['Overcast', 'Hot', 'Normal', 'Weak', 'Yes'],
    ['Rainy', 'Mild', 'High', 'Strong', 'No']
]

# Separate features and target variable
features = [row[:-1] for row in data]
target = [row[-1] for row in data]

# Convert categorical data to numerical data using one-hot encoding
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
features_encoded = encoder.fit_transform(features)

# Initialize a decision tree classifier
clf = tree.DecisionTreeClassifier()

# Train the classifier
clf = clf.fit(features_encoded, target)

# Visualize the decision tree
plt.figure(figsize=(10, 6))
tree.plot_tree(clf, feature_names=encoder.get_feature_names_out())
plt.show()

# Table representation of the decision tree
headers = encoder.get_feature_names_out()
tree_table = tabulate(clf.tree_.value, headers, tablefmt='pretty')
print(f"Decision Tree Table:\n{tree_table}")

# Make predictions for a new data point
new_data_point = [['Sunny', 'Hot', 'High', 'Strong']]  # Make sure to use a list of lists
new_data_point_encoded = encoder.transform(new_data_point)
prediction = clf.predict(new_data_point_encoded)

# Print the prediction
print(f'The model predicts: {prediction[0]}')
```
