```python
from copy import deepcopy
import time
import random
import csv

class Node:
    def __init__(self, board, move=None, parent=None, value=None):
        self.board = board
        self.value = value
        self.move = move
        self.parent = parent

    def get_children(self, is_computer_1):
        current_state = deepcopy(self.board)
        available_moves = []
        children_states = []
        big_letter = ""
        queen_row = 0
        if is_computer_1 is True:
            available_moves = Checkers.find_computer1_available_moves(current_state)
            big_letter = "C"
            queen_row = 7
        else:
            available_moves = Checkers.find_computer2_available_moves(current_state)
            big_letter = "B"
            queen_row = 0
        for i in range(len(available_moves)):
            old_i = available_moves[i][0]
            old_j = available_moves[i][1]
            new_i = available_moves[i][2]
            new_j = available_moves[i][3]
            state = deepcopy(current_state)
            Checkers.make_a_move(state, old_i, old_j, new_i, new_j, big_letter, queen_row)
            children_states.append(Node(state, [old_i, old_j, new_i, new_j]))
        return children_states

    def get_board(self):
        return self.board


class Checkers:
    def __init__(self):
        self.matrix = [[], [], [], [], [], [], [], []]
        self.computer1_pieces = 12
        self.computer2_pieces = 12
        self.available_moves = []
        self.mandatory_jumping = True
        for row in self.matrix:
            for i in range(8):
                row.append("---")
        self.position_computer1()
        self.position_computer2()

    def position_computer1(self):
        for i in range(3):
            for j in range(8):
                if (i + j) % 2 == 1:
                    self.matrix[i][j] = ("c" + str(i) + str(j))

    def position_computer2(self):
        for i in range(5, 8, 1):
            for j in range(8):
                if (i + j) % 2 == 1:
                    self.matrix[i][j] = ("b" + str(i) + str(j))

    def print_matrix(self):
        i = 0
        print()
        for row in self.matrix:
            print(i, end="  |")
            i += 1
            for elem in row:
                print(elem, end=" ")
            print()
        print()
        for j in range(8):
            if j == 0:
                j = "     0"
            print(j, end="   ")
        print("\n")

    def find_computer1_available_moves(board):
        available_moves = []
        available_jumps = []
        for m in range(8):
            for n in range(8):
                if board[m][n][0] == "c":
                    if Checkers.check_computer1_moves(board, m, n, m + 1, n + 1):
                        available_moves.append([m, n, m + 1, n + 1])
                    if Checkers.check_computer1_moves(board, m, n, m + 1, n - 1):
                        available_moves.append([m, n, m + 1, n - 1])
                    if Checkers.check_computer1_jumps(board, m, n, m + 1, n - 1, m + 2, n - 2):
                        available_jumps.append([m, n, m + 2, n - 2])
                    if Checkers.check_computer1_jumps(board, m, n, m + 1, n + 1, m + 2, n + 2):
                        available_jumps.append([m, n, m + 2, n + 2])
                elif board[m][n][0] == "C":
                    if Checkers.check_computer1_moves(board, m, n, m + 1, n + 1):
                        available_moves.append([m, n, m + 1, n + 1])
                    if Checkers.check_computer1_moves(board, m, n, m + 1, n - 1):
                        available_moves.append([m, n, m + 1, n - 1])
                    if Checkers.check_computer1_moves(board, m, n, m - 1, n - 1):
                        available_moves.append([m, n, m - 1, n - 1])
                    if Checkers.check_computer1_moves(board, m, n, m - 1, n + 1):
                        available_moves.append([m, n, m - 1, n + 1])
                    if Checkers.check_computer1_jumps(board, m, n, m + 1, n - 1, m + 2, n - 2):
                        available_jumps.append([m, n, m + 2, n - 2])
                    if Checkers.check_computer1_jumps(board, m, n, m - 1, n - 1, m - 2, n - 2):
                        available_jumps.append([m, n, m - 2, n - 2])
                    if Checkers.check_computer1_jumps(board, m, n, m - 1, n + 1, m - 2, n + 2):
                        available_jumps.append([m, n, m - 2, n + 2])
                    if Checkers.check_computer1_jumps(board, m, n, m + 1, n + 1, m + 2, n + 2):
                        available_jumps.append([m, n, m + 2, n + 2])

        if len(available_jumps) == 0:
            return available_moves
        else:
            return available_jumps

    def check_computer1_jumps(board, old_i, old_j, via_i, via_j, new_i, new_j):
        if new_i > 7 or new_i < 0:
            return False
        if new_j > 7 or new_j < 0:
            return False
        if board[via_i][via_j] == "---":
            return False
        if board[via_i][via_j][0] == "C" or board[via_i][via_j][0] == "c":
            return False
        if board[new_i][new_j] != "---":
            return False
        if board[old_i][old_j] == "---":
            return False
        if board[old_i][old_j][0] == "b" or board[old_i][old_j][0] == "B":
            return False
        return True

    def check_computer1_moves(board, old_i, old_j, new_i, new_j):
        if new_i > 7 or new_i < 0:
            return False
        if new_j > 7 or new_j < 0:
            return False
        if board[old_i][old_j] == "---":
            return False
        if board[new_i][new_j] != "---":
            return False
        if board[old_i][old_j][0] == "b" or board[old_i][old_j][0] == "B":
            return False
        if board[new_i][new_j] == "---":
            return True

    def find_computer2_available_moves(board):
        available_moves = []
        available_jumps = []
        for m in range(8):
            for n in range(8):
                if board[m][n][0] == "b":
                    if Checkers.check_computer2_moves(board, m, n, m - 1, n - 1):
                        available_moves.append([m, n, m - 1, n - 1])
                    if Checkers.check_computer2_moves(board, m, n, m - 1, n + 1):
                        available_moves.append([m, n, m - 1, n + 1])
                    if Checkers.check_computer2_jumps(board, m, n, m - 1, n - 1, m - 2, n - 2):
                        available_jumps.append([m, n, m - 2, n - 2])
                    if Checkers.check_computer2_jumps(board, m, n, m - 1, n + 1, m - 2, n + 2):
                        available_jumps.append([m, n, m - 2, n + 2])
                elif board[m][n][0] == "B":
                    if Checkers.check_computer2_moves(board, m, n, m - 1, n - 1):
                        available_moves.append([m, n, m - 1, n - 1])
                    if Checkers.check_computer2_moves(board, m, n, m - 1, n + 1):
                        available_moves.append([m, n, m - 1, n + 1])
                    if Checkers.check_computer2_jumps(board, m, n, m - 1, n - 1, m - 2, n - 2):
                        available_jumps.append([m, n, m - 2, n - 2])
                    if Checkers.check_computer2_jumps(board, m, n, m - 1, n + 1, m - 2, n + 2):
                        available_jumps.append([m, n, m - 2, n + 2])
                    if Checkers.check_computer2_moves(board, m, n, m + 1, n - 1):
                        available_moves.append([m, n, m + 1, n - 1])
                    if Checkers.check_computer2_jumps(board, m, n, m + 1, n - 1, m + 2, n - 2):
                        available_jumps.append([m, n, m + 2, n - 2])
                    if Checkers.check_computer2_moves(board, m, n, m + 1, n + 1):
                        available_moves.append([m, n, m + 1, n + 1])
                    if Checkers.check_computer2_jumps(board, m, n, m + 1, n + 1, m + 2, n + 2):
                        available_jumps.append([m, n, m + 2, n + 2])
        if len(available_jumps) == 0:
            return available_moves
        else:
            return available_jumps

    def check_computer2_moves(board, old_i, old_j, new_i, new_j):
        if new_i > 7 or new_i < 0:
            return False
        if new_j > 7 or new_j < 0:
            return False
        if board[old_i][old_j] == "---":
            return False
        if board[new_i][new_j] != "---":
            return False
        if board[old_i][old_j][0] == "c" or board[old_i][old_j][0] == "C":
            return False
        if board[new_i][new_j] == "---":
            return True

    def check_computer2_jumps(board, old_i, old_j, via_i, via_j, new_i, new_j):
        if new_i > 7 or new_i < 0:
            return False
        if new_j > 7 or new_j < 0:
            return False
        if board[via_i][via_j] == "---":
            return False
        if board[via_i][via_j][0] == "B" or board[via_i][via_j][0] == "b":
            return False
        if board[new_i][new_j] != "---":
            return False
        if board[old_i][old_j] == "---":
            return False
        if board[old_i][old_j][0] == "c" or board[old_i][old_j][0] == "C":
            return False
        return True

    def evaluate_states(self, is_computer_1):
      t1 = time.time()
      current_state = Node(deepcopy(self.matrix))
      first_computer_moves = current_state.get_children(is_computer_1)
      if len(first_computer_moves) == 0:
          if self.computer2_pieces > self.computer1_pieces:
              print("Computer " + ("1" if is_computer_1 else "2") + " has no available moves left, and the other computer has more pieces left.\nComputer " + ("1" if not is_computer_1 else "2") + " wins!")
          else:
              print("Computer " + ("1" if is_computer_1 else "2") + " has no available moves left. and the other computer has less pieces left.\nComputer " + ("1" if is_computer_1 else "2") + " wins!")
          exit()

      if is_computer_1:
          print("Computer 1's turn.")
      else:
          print("Computer 2's turn.")
      print("Thinking...")

      random_move = random.choice(first_computer_moves)
      new_board = random_move.get_board()
      move = random_move.move
      self.matrix = new_board
      t2 = time.time()
      diff = t2 - t1
      print("Computer " + ("1" if is_computer_1 else "2") + " has moved (" + str(move[0]) + "," + str(move[1]) + ") to (" + str(move[2]) + "," + str(move[3]) + ").")
      print("It took him " + str(diff) + " seconds.")

    def make_a_move(board, old_i, old_j, new_i, new_j, big_letter, queen_row):
        letter = board[old_i][old_j][0]
        i_difference = old_i - new_i
        j_difference = old_j - new_j
        if i_difference == -2 and j_difference == 2:
            board[old_i + 1][old_j - 1] = "---"
        elif i_difference == 2 and j_difference == 2:
            board[old_i - 1][old_j - 1] = "---"
        elif i_difference == 2 and j_difference == -2:
            board[old_i - 1][old_j + 1] = "---"
        elif i_difference == -2 and j_difference == -2:
            board[old_i + 1][old_j + 1] = "---"
        if new_i == queen_row:
            letter = big_letter
        board[old_i][old_j] = "---"
        board[new_i][new_j] = letter + str(new_i) + str(new_j)

    def matrix_to_string(self):
        string = ""
        for row in self.matrix:
            for elem in row:
                string += elem
        return string

    def play(self,x):
      c1 = []
      c2 = []
      while True:
          self.evaluate_states(True)
          self.print_matrix()
          c1.append(self.matrix_to_string())

          if len(Checkers.find_computer2_available_moves(self.matrix)) == 0:
                print("Computer 2 has no available moves left.\nComputer 1 wins!")
                for i in range(len(c1)):
                    if(x.get(c1[i]) == None):
                        x[c1[i]] = [100,1]
                    else:
                        a = x[c1[i]]
                        x[c1[i]] = [(a[0]+100)/(a[1]+1),a[1]+1]

                for i in range(len(c2)):
                    if(x.get(c2[i]) == None):
                        x[c2[i]] = [-100,1]
                    else:
                        a = x[c2[i]]
                        x[c2[i]] = [(a[0]-100)/(a[1]+1),a[1]+1]
                break

          self.evaluate_states(False)
          self.print_matrix()
          c2.append(self.matrix_to_string())

          if len(Checkers.find_computer1_available_moves(self.matrix)) == 0:
                print("Computer 1 has no available moves left.\nComputer 2 wins!")
                for i in range(len(c1)):
                    if(x.get(c1[i]) == None):
                        x[c1[i]] = [-100,1]
                    else:
                        a = x[c1[i]]
                        x[c1[i]] = [(a[0]-100)/(a[1]+1),a[1]+1]

                for i in range(len(c2)):
                    if(x.get(c2[i]) == None):
                        x[c2[i]] = [100,1]
                    else:
                        a = x[c2[i]]
                        x[c2[i]] = [(a[0]+100)/(a[1]+1),a[1]+1]
                break

      return x


if __name__ == '__main__':
    x = dict()
    with open("/content/checkers.csv", "r") as file:
        a = csv.reader(file)
        for row in a:
            x[row[0]] = row[1]
    checker = Checkers()
    checker.play(x)
    with open("/content/checkers.csv", "w") as file:
        a = csv.writer(file)
        for i in x:
            a.writerow([i,x[i][0],x[i][1]])
```
