```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
image = plt.imread("/content/IMG_20221225_194950.jpg")
print(image.shape)
w, h, c = image.shape

g_image = np.uint8(np.zeros((w,h)))
bnr_image = np.uint8(np.zeros((w,h)))
thr_image = np.uint8(np.zeros((w,h)))

for i in range(w):
  for j in range(h):
    g_image[i,j]=((0.0722*image[i,j,0])+(0.7152*image[i,j,1])+(0.2126*image[i,j,2]))

for i in range(w):
  for j in range(h):
      if g_image[i,j]>128:
        bnr_image[i,j]=255
      else:
        bnr_image[i,j]=0

total = 0
for i in range(w):
  for j in range(h):
      total+=g_image[i,j]

thresold = total/(w*h)

for i in range(w):
  for j in range(h):
      if g_image[i,j]>thresold:
        thr_image[i,j]=255
      else:
        thr_image[i,j]=0

fig = plt.figure(figsize=(10,10))
columns = 2
rows = 2

fig.add_subplot(rows,columns,1)
plt.imshow(image)
plt.axis("off")
plt.title("color image")

fig.add_subplot(rows,columns,2)
plt.imshow(g_image,"gray")
plt.axis("off")
plt.title("grey image")

fig.add_subplot(rows,columns,3)
plt.imshow(bnr_image,"gray")
plt.axis("off")
plt.title("binary image")

fig.add_subplot(rows,columns,4)
plt.imshow(thr_image,"gray")
plt.axis("off")
plt.title("Thresold image")
```

```python
red_image = np.uint8(np.zeros((600,600,3))) 
red_image.shape
for i in range(600):
  for j in range(600):
    red_image[i,j,0] =255

grn_image = red_image.copy()
for i in range(600):
  for j in range(600):
    grn_image[i,j,0] = 0
    grn_image[i,j,1] = 255

fig = plt.figure(figsize=(10,10))
c = 2
r =1

fig.add_subplot(r,c,1)
plt.imshow(red_image)
plt.axis("off")
plt.title("original")

fig.add_subplot(r,c,2)
plt.imshow(grn_image)
plt.axis("off")
plt.title("green")
```
