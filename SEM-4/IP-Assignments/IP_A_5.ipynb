```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = plt.imread("Fig0221(a)(ctskull-256).tif")
img.shape
```

```python
plt.imshow(img,"gray")
```

```python
def IntensityResolution(level):
    bits = np.log2(level)
    new_img = np.floor(img/ (2**(8 - bits)))
    return new_img
```

```python
fig = plt.figure(figsize=(20,20))
c,r = 2,4
fig.add_subplot(r,c,1)
plt.imshow(IntensityResolution(256),"gray")
plt.axis("off")
plt.title("256 level")
fig.add_subplot(r,c,2)
plt.imshow(IntensityResolution(128),"gray")
plt.axis("off")
plt.title("128 level")
fig.add_subplot(r,c,3)
plt.imshow(IntensityResolution(64),"gray")
plt.axis("off")
plt.title("64 level")
fig.add_subplot(r,c,4)
plt.imshow(IntensityResolution(32),"gray")
plt.axis("off")
plt.title("28 level")
fig.add_subplot(r,c,5)
plt.imshow(IntensityResolution(16),"gray")
plt.axis("off")
plt.title("16 level")
fig.add_subplot(r,c,6)
plt.imshow(IntensityResolution(8),"gray")
plt.axis("off")
plt.title("8 level")
fig.add_subplot(r,c,7)
plt.imshow(IntensityResolution(4),"gray")
plt.axis("off")
plt.title("4 level")
fig.add_subplot(r,c,8)
plt.imshow(IntensityResolution(2),"gray")
plt.axis("off")
plt.title("2 level")
```
